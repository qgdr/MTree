```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')
x, y, z = np.indices([2, 2, 2])
filled = np.ones([1, 1, 1])
ax.voxels(x, y, z, filled=filled, alpha=0.3, edgecolor='k')

verts = [[(1.5, 0, 0), (0, 1.5, 0), (0, 0, 1.5)]]
ax.add_collection3d(Poly3DCollection(verts, facecolors='y', alpha=0.3))


verts2 = [[(1, 0, 0.5), (1, 0.5, 0), (0.5, 1, 0), (0, 1, 0.5), (0, 0.5, 1), (0.5, 0, 1)]]
ax.add_collection3d(Poly3DCollection(verts2, facecolors='r', edgecolors='k', alpha=0.5))

ax.view_init(20, 55)
# ax.set_axis_off()
ax.set_xlim3d(0.04, 1.6)
ax.set_ylim3d(0.04, 1.6)
ax.set_zlim3d(0.04, 1.6)
plt.savefig('SparseConvexHull.png')
plt.show()
```
